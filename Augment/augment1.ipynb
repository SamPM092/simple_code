```python
import imgaug.augmenters as iaa
import numpy as np
import matplotlib.pyplot as plt
import imageio.v2 as imageio
from PIL import Image
```

```python
# Load an example image
image_path = "citrus/canker/1.png.png"
image = imageio.imread(image_path)
size = image.size

# Define data augmentation pipeline
augmentation_pipeline = iaa.Sequential([
    iaa.Fliplr(0.5),        # Flip horizontally with a 50% chance
    iaa.Affine(rotate=(-180, 180)),  # Rotate between -25 and 25 degrees
    iaa.GaussianBlur(sigma=(0, 1.0))  # Apply Gaussian blur with a sigma between 0 and 1.0
])

```

```python
# Generate augmented images
augmented_images = augmentation_pipeline(images=np.array([image]))
```

```python
# Display original and augmented images
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(augmented_images[0])
plt.title("Augmented Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
